```python
### Univariate Linear Regression for 2 different X variables -- Population and Skill Group and y variable - drainOrGain
```

```python
import os
import re
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split

import warnings
warnings.filterwarnings("ignore")
```

```python
data1 = pd.read_csv('../data/literacy_rate_public.csv')
print(f"Shape of the data is: {data1.shape}")
data1.head()
```

```python
data1.set_index('country')
data1.head()
```

```python
data2 = pd.read_csv('../data/skill_migration_public.csv')
print(f"Shape of the data is: {data2.shape}")
data2.head()
```

```python
data2.drop_duplicates(subset=['country'])
data2.shape
```

```python
data3 = pd.read_csv('../data/gdp_public.csv')
print(f"Shape of the data is: {data2.shape}")
data3.head()
```

```python
combined_data1 = pd.merge(data2, data1, on='country', how='inner')
combined_data1
```

```python
combined_data = pd.merge(combined_data1, data3, on='country', how='inner')
combined_data
```

```python
combined_data.dtypes
```

```python
combined_data['drainOrGain'] = combined_data['population'] * combined_data['Delta']/10000
```

```python
clean_data = pd.get_dummies(combined_data, columns=['skill_group'])
clean_data.head()
```

```python
clean_data['country']
```

```python
y = clean_data["Delta"].values.reshape(-1,1)

X = clean_data.drop(['skill_group_id', 'country_code'], axis=1)

print(X.shape, y.shape)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(n_jobs=100)
```

```python
clean_data.dtypes
```

```python
model.fit(X, y)
plt.scatter(X, y)
plt.plot(X, model.predict(X), c='red')
```
