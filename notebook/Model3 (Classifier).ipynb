```python
import os
import re
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split

import warnings
warnings.filterwarnings("ignore")
```

```python
data1 = pd.read_csv('../data/literacy_rate_public.csv')
print(f"Shape of the data is: {data1.shape}")
data1.head()
```

```python
data1.set_index('country')
data1.head()
```

```python
data2 = pd.read_csv('../data/skill_migration_public.csv')
print(f"Shape of the data is: {data2.shape}")
data2.head()
```

```python
data2.set_index('country')
```

```python
data3 = pd.read_csv('../data/gdp_public.csv')
print(f"Shape of the data is: {data3.shape}")
data3.head()
```

```python
combined_data1 = pd.merge(data2, data1, on='country', how='inner')
combined_data1
```

```python
combined_data = pd.merge(combined_data1, data3, on='country', how='inner')
combined_data
```

```python
combined_data.dtypes
```

```python
combined_data['drainOrGain'] = combined_data['Delta']/100
```

```python
combined_data['drainOrGain'] = combined_data['drainOrGain'].astype(int) 
```

```python
combined_data.head()
```

```python
# store the target variable in y and everything else goes in X
y = combined_data['drainOrGain']

# Encoding and Dummying 
clean_data = pd.get_dummies(combined_data, columns=['skill_group'])
# we're also dropping several fields because it adds no value to the prediction
X = clean_data.drop(['country_code','skill_group_id', 'skill', 'country', 'income_group', 'world_region'], axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=3)

X_train.shape 
```

```python
X_train.dtypes
```

```python
y
```

```python
X_train.isnull().sum()
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()

# param_grid = {"n_estimators" : [10, 20, 50, 100],
#              "max_depth" : [None, 6, 8, 10],
#              "max_leaf_nodes": [None, 5, 10, 20], 
#              "min_impurity_split": [0.1, 0.2, 0.3, 0.4]}

param_grid = {"n_estimators" : [10, 20, 100, 1000],
             "max_depth" : [None, 6, 10],
             "max_leaf_nodes": [None, 5,  20], 
             "min_impurity_split": [0.1, 0.2,  0.4]}
```

```python
from sklearn.model_selection import GridSearchCV

grid = GridSearchCV(model, param_grid=param_grid, cv=3, verbose=2)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
import pickle

with open('../models/RFC_model.pkl','wb') as f:
    pickle.dump(grid, f)
```

```python
with open('RFC_model.pkl', 'rb') as f:
    model = pickle.load(f)
```

```python
print(f"Test set score: {model.score(X_test, y_test)}")
```
